```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
y=x*3
```

```python
x
```

```python
z=x*5
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=4)
plt.grid(True)
plt.title('andekhi anjami si')
plt.ylabel("height in inch")
plt.xlabel("age in year")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,linestyle='-',linewidth=4,color='green')

plt.subplot(1,2,2)
plt.plot(x,z,linestyle=':',linewidth=6,color='pink')
```

*scatterplot

```python
x=np.arange(0,50,3)
```

```python
z=x*2
```

```python
y=np.arange(0,100,6)
```

```python
from random import shuffle
shuffle(x)
shuffle(y)
shuffle(z)
```

```python
plt.scatter(x,y,marker='*',s=200,c='g')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",c='g',s=200)

plt.subplot(1,2,2)
plt.scatter(x,z,marker=".",c="r",s=100)
plt.show()
```

*boxplot

```python
x=[2,2,3,4,5,4,3,2]
y=[1,3,4,6,2,1,5,7]
z=[2,5,6,7,2,3,1,0]
data=[x,y,z]
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
fruit=['apple','mango','orange','guava']
quantity=[23,34,27,16]
plt.pie(quantity,labels=fruit,autopct='%0.1f%%')
plt.show()
```

*doughnut chart

```python
plt.pie(quantity,labels=fruit,radius=1)
plt.pie([10],radius=0.30,colors='white')
plt.show()
```
